Zadanie należy wykonać wykorzystując program `BagOfWords` z poprzednich zajęć.


Przerób ją tak, żeby interpunkcja, cyfry i wszelkie inne znaki nie przeszkadzały w parsowaniu tekstu. Uruchom ja na [tekście hamleta](http://www.gutenberg.org/cache/epub/1787/pg1787.txt). Ile razy występuje słowo hamlet? Jak brzmi 10 najczęściej występujących słów?

```python
# solutions for exerciese 2 and 5
from collections import defaultdict
import json

from collections import defaultdict
import string

class BagOfWords:
    def __init__ (self, text):
        self.words = defaultdict(int)
        if type(text) == str:
            text = ''.join(i for i in text if not i.isdigit())
            text = text.translate(str.maketrans('', '', string.punctuation))
            text = text.lower()
            for word in text.split():
                self.words[word] += 1
        else:
            with text as lines:
                for line in lines:
                    text = ''.join(i for i in text if not i.isdigit())
                    text = text.translate(str.maketrans('', '', string.punctuation))
                    text = text.lower()
                    for word in text.split():
                        self.words[word] += 1

    def __str__ (self):
        return_string = ""
        for key, value in sorted(self.words.items(), key = lambda x: x[1], reverse = True):
            return_string += key + ":" + str(value) + ", "
        return return_string

    def __iter__(self):
        return iter(dict(sorted(self.words.items(), key = lambda x: x[1], reverse = True)))

    def __add__(self, d):
        bag = BagOfWords("")
        for key, value in d.words.items():
            bag.words[key] += value
        for key, value in self.words.items():
            bag.words[key] += value
        return bag

    def __getitem__ (self, name):
        return self.words[name]
    
    def __setitem__ (self, name, value):
        self.words[name] = value
    def topTen(self):
        return_string = ""
        for key, value in sorted(self.words.items(), key=lambda x: x[1], reverse=True)[:10]:
            return_string += key + ":" + str(value) + ", "
        return return_string
    def load(self, json_name):
        with open(json_name) as json_file:
            data = json.load(json_file)
        print(data)
        new_bow = BagOfWords("")
        new_bow.words = data
        return new_bow
        
    def save(self, json_name):
        with open(json_name, "w") as json_file:
            json.dump(self.words, json_file, indent=4)

bowHamlet = BagOfWords(open("Hamlet.rtf"))

print(bowHamlet.topTen())
print(bowHamlet["hamlet"])
bowHamlet.save("hamlet_bow.json")
bow = BagOfWords("Iza z matiza Agata z fiata")
bow.save("bow.json")
bow_loaded = bow.load("bow.json")
print(bow)
print(bow_loaded)
```

## Zadanie 4

Wykorzystując [https://gist.github.com/pamelafox/986163](https://gist.github.com/pamelafox/986163) podaj aktualną godzinę we:
- wszystkich krajach, wyświetlając je zgrupowane względem kontynentów,
- przeprowadź symulacyjne wyświetlanie kolejnych krajów w miarę jak w danym kraju wybija północ, opóżnienie wyświetlania ustaw proporcjonalne do realnego czasu

```python
import countryinfo
import pprint
import pytz
from datetime import datetime
grouped_countries = {}

def get_current_time(timezone_name):
    timezone = pytz.timezone(timezone_name)
    return datetime.now(timezone)

for country in countryinfo.countries:
    continent = country['continent']
    if continent not in grouped_countries:
        grouped_countries[continent] = []
    grouped_countries[continent].append(country)

for continent, countries_list in grouped_countries.items():
    print(f"{continent} - ", end="")
    for country in countries_list:
        pprint.pprint(f"{country['name']} {get_current_time(country['timezones'][0])}, ", indent = 4)
```
